## Barcelona

```python
import os
import requests
import chardet
import pandas as pd

endpoint = "https://opendata-ajuntament.barcelona.cat/data/dataset/2767159c-1c98-46b8-a686-2b25b40cb053/resource/3abb2414-1ee0-446e-9c25-380e938adb73/download"
filepath = os.path.join(folder, 'spain_activity_barc.csv')
res = requests.get(endpoint)
with open(filepath, 'wb') as f:
    f.write(res.content)
with open(filepath, 'rb') as f:
    rawdata = f.read()
    guess = chardet.detect(rawdata)
encoding_to_use = guess['encoding'] if guess['encoding'] else 'utf-8-sig'
```

```python
df = pd.read_csv(filepath, sep=';', encoding=encoding_to_use, engine='python', on_bad_lines='skip')
df.head(10)
```

```python
endpoint = "https://opendata-ajuntament.barcelona.cat/data/dataset/9cefbfa2-bcdf-44a0-b63a-372b48f9da93/resource/fc0a920c-79e5-43c2-8b7a-fe29296a2f90/download/2024_2s_data_set_opendata_terrasses.csv"
filepath = os.path.join(folder, 'spain_shops_barc.csv')
res = requests.get(endpoint)
with open(filepath, 'wb') as f:
    f.write(res.content)
with open(filepath, 'rb') as f:
    rawdata = f.read()
    guess = chardet.detect(rawdata)
encoding_to_use = guess['encoding'] if guess['encoding'] else 'utf-8-sig'
```

```python
df = pd.read_csv(filepath, sep=';', encoding=encoding_to_use, engine='python', on_bad_lines='skip')
df.head(10)
```
